```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset=pd.read_csv(r"C:\Users\Mirza\Desktop\Iris.csv")
dataset.head()
```

```python
dataset['Species'].unique()
dataset.info()
```

```python
dataset.describe()
```

```python
x=dataset[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
y=dataset['Species'].values
```

```python
print(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print('Accuracy is',accuracy_score(pred,y_test))
print('\nClassification: \n ',classification_report(pred,y_test))
```

```python
from sklearn import tree
feature=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
plt.figure(figsize=(30,20))
tree.plot_tree(dt.fit(x,y),feature_names=feature,filled=True,precision=3,proportion=True,rounded=True)
plt.show()
```
